```python
# Kublar (Cubes):
# Ədədin üçüncü dərəcə qüvvəti kub adlanır. Məsələn,
# 2-nin kubu Python-da 2**3 kimi yazılır. İlk 10 kubun siyahısını yaradın (yəni,
# 1-dən 10-a qədər hər bir tam ədədin kubu) və hər bir kubun dəyərini çap etmək üçün
# "for" döngüsündən istifadə edin.


cubes = []

for i in range(1,11):
    cube = i**3
    cubes.append(cube)
    
print(cubes)
```
