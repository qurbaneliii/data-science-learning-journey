```python
# Add two decimal numbers and separate integer and decimal part
x = 6.5
y = 3.2
cəm = x + y
tam_hissə = int(cəm)   # Extract the integer part
kəsr_hissə = cəm - tam_hissə  # Extract the decimal part
print(tam_hissə)
print(kəsr_hissə)
```

```python
# Round the decimal part to 2 digits
x = 6.5
y = 3.2
cəm = x + y
tam_hissə = int(cəm)
kəsr_hissə = round(cəm - tam_hissə, 2)
print(tam_hissə)
print(kəsr_hissə)
```
