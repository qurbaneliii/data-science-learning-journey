```python
grades = [95, 72, 69, 50, 88]
letters = []
for i in grades:
    if 90<i<100:
        letters.append("A")
    elif 80<i<90:
        letters.append("B")
    elif 70<i<80:
        letters.append("C")
    elif 60<i<70:
        letters.append("D")
    else:
        letters.append("F")
print(letters)
```

```python
# This is a simple login system project.
# The program allows a user to attempt to login up to 3 times.
# If the correct username and password are entered, login is successful.
# If the user fails 3 times, the login is blocked.

correct_username = "admin"
correct_password = "1234"

success = False  # Flag to check whether login was successful

for i in range(3):
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == correct_username and password == correct_password:
        print("Login successful")  # User entered correct credentials
        success = True
        break  # Exit the loop since login succeeded
    else:
        # Show remaining attempts
        print(f"Incorrect credentials! {2 - i} attempts left.")

if success:
    pass  # Login succeeded, nothing else to do
else:
    print("Login blocked! Too many failed attempts")  # User failed 3 times
```
